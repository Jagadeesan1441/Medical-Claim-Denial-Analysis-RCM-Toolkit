```python

import pandas as pd
import random
from faker import Faker

fake = Faker()


cpt_codes = [
    '99213', '99214', '99203', '93000', '36415', '81001','90471','20550', '20610', '71020',
    '74177', '80053','87635', '99285','99291','29580','96372','11721','99281','99232']


insurance_companies = [
    'Aetna', 'UnitedHealthcare', 'Blue Cross Blue Shield', 'Cigna', 'Humana',
    'Anthem', 'Kaiser Permanente', 'Molina Healthcare', 'Centene Corporation', 'Medicare'
]


physicians = [fake.name() for _ in range(30)]

denial_reasons = [
    'Modifier issue', 'LCD/NCD mismatch', 'Bundling edits (NCCI)', 'Lack of documentation',
    'Prior authorization missing', 'Credentialing issue', 'Duplicate claim',
    'Invalid CPT code', 'Late submission', 'Insurance coverage expired'
]


data = []

for _ in range(5000):
    cpt = random.choice(cpt_codes)
    payer = random.choice(insurance_companies)
    physician = random.choice(physicians)

    # Simulate payment and balance (some denied = balance > 0)
    # Assume ~30% denial rate
    denied = random.choices([True, False], weights=[30, 70])[0]
    payment = 0 if denied else round(random.uniform(50, 500), 2)
    balance = 0 if not denied else round(random.uniform(50, 500), 2)

    denial_reason = random.choice(denial_reasons) if denied else ''

    data.append({
        'CPT Code': cpt,
        'Insurance Company': payer,
        'Physician': physician,
        'Payment': payment,
        'Balance': balance,
        'Denial Reason': denial_reason
    })


df = pd.DataFrame(data)
df.to_csv('newsynthetic_medical_billing.csv', index=False)

```

#3)RECOMMEND FIX AND STRATEGIES



 1. Modifier Issues

Fix: Provide refresher training for coding teams on common modifiers (e.g., -25, -59) and payer-specific rules.

Strategy: Implement pre-submission claim edits in the billing system to flag missing or incorrect modifiers.

Workflow Recommendation: Embed a coding checklist or smart prompts in your EHR to reduce modifier-related errors.






 2. Lack of Documentation

Fix: Standardize documentation templates for common procedures and diagnoses to ensure medical necessity is clearly supported.

Strategy: Create a feedback loop between coders and providers when documentation is flagged as insufficient.

Workflow Recommendation: Train front-line clinicians on what documentation is critical for billing compliance.





 3. LCD/NCD Mismatch

Fix: Build LCD/NCD checks into your coding software or EMR to alert staff before claim submission.

Strategy: Educate providers on Medicare coverage rules for high-denial services.

Workflow Recommendation: Use cheat sheets or quick-reference tools for complex services with frequent coverage limits.





 4. Prior Authorization Missing

Fix: Automate prior auth checks at the scheduling stage using payer portals or clearinghouse tools.

Strategy: Assign a pre-auth coordinator or establish a prior authorization team.

Workflow Recommendation: Ensure front desk staff verify and obtain prior auth before scheduling elective services.





 5. Credentialing or Provider Enrollment Issues

Fix: Maintain a central credentialing calendar with expiration dates and application status.

Strategy: Set automated reminders 90/60/30 days before recredentialing is due.

Workflow Recommendation: Create a provider onboarding checklist that includes timely payer enrollment submissions.





 6. Duplicate Claims & Late Submissions

Fix: Strengthen your claim tracking system to monitor submission status and prevent resubmission without cause.

Strategy: Automate claim submission within 24 hours of charge entry.

Workflow Recommendation: Educate billing staff on the correct use of rejections vs. denials to avoid unnecessary resubmissions.
